Bubble chart is extension of scatter plot. Here is nice creative example how to visualise 4D information into the 2D graph. Coordinates of cities are x and y, area (size) of the city is size of the bubble and aplha channel of bubble is used for visualise population density.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
data = {
    'City': ['New York', 'Tokyo', 'London', 'Mumbai', 'Sydney', 'Rio de Janeiro'],
    'Latitude': [40.7128, 35.6895, 51.5099, 19.0760, -33.8688, -22.9094],
    'Longitude': [-74.0060, 139.6917, -0.1180, 72.8777, 151.2093, -43.215],
    'Area': [784, 2194, 1572, 603, 12144, 1182], # kilometers squared
    'Population': [1984000, 1396000, 8982000, 21673000, 5312000, 6748000]
}

cities_df = pd.DataFrame(data)

# Calculate population densities
densities = cities_df['Population'] / cities_df['Area']

# Normalize Population Density for alpha values
alpha_values = densities / densities.max()

# Plotting the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(
    cities_df['Longitude'],
    cities_df['Latitude'],
    s=cities_df['Area'],
    c='blue',  # color of bubbles
    alpha=alpha_values,  # transparency of bubbles
    edgecolors='black',  # color of bubble edges
)

# Adding labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Bubble Chart of Cities')

# Adding city names as annotations on their coordinates
for i, city in enumerate(cities_df['City']):
    plt.annotate(city, (cities_df['Longitude'][i], cities_df['Latitude'][i]), ha='center', va='center')

plt.show()
```

As you can see, Sydney is biggest, but has proportionally smallest population, so its almost invisible (cause of extremely low alpha value).

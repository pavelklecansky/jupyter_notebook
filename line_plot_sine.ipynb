Visualize sine wave (high resolution)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data (100 points)
x = np.linspace(0, 2 * np.pi, 100) # 0-2*pi
y = np.sin(x)

# Create a plot
plt.plot(x, y, label='Sin(x)')

# Customize the plot
plt.title('Sine Wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.grid()
plt.legend()

# Show the plot
plt.show()
```

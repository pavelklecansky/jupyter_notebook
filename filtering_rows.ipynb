```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Adam'],
    'Age': [25, 30, 35, 15],
    'Height': [170, 175, 183, 165],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Washington']
}
df = pd.DataFrame(data)

# Filtering rows based on a condition
filtered_df = df[df['Age'] > 29]
print("Filtered older than 29:")
print(filtered_df)
print("\n")

# You can combine multiple conditions
# using logical operators (&, |, ~)
filtered_df = df[(df['Age'] > 29) & (df['City'] == 'Los Angeles')]
print("Filtered older than 29 and from Los Angeles:")
print(filtered_df)
print("\n")

# You can combine filter by first letter of name
filtered_name = df[df['Name'].str.startswith('A')]
print("Filtered people with name starting with A letter:")
print(filtered_name)
print("\n")

# Or filter by enumeration of values
print("Filtered people with names Alice or Bob:")
filtered_names = df[df['Name'].isin(['Alice', 'Bob'])]
print(filtered_df)
print("\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with three clusters representing women, men and children
X, _ = make_blobs(n_samples=300, centers=[(160, 55),(175, 70),(130, 30)],
       cluster_std=[8, 10, 5], random_state=42)

# Create a KMeans model with three clusters
kmeans = KMeans(n_clusters=3, random_state=42)
# Fit the model to the data
kmeans.fit(X)
# Get cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Get the inertia
inertia = kmeans.inertia_
print("Inertia:", inertia)
# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette avg:", silhouette_avg)

# Plot the original data points
plt.figure(figsize=(10, 6))
# Plot the cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, marker='X', label='Cluster Centers (Women, Men, Children)')
# Plot the clustered data points with different colors for each cluster
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, edgecolors='k', alpha=0.7, label='Clustered Data')
# Set labels and legend
plt.title('K-Means Clustering of Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
#plt.legend()
plt.grid()

# Show the plot
plt.show()
```

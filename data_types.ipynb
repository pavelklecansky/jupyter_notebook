Here is list of common **data types** in Python:

```python
# Integer
x = 10
print('x:', x)

# Float
pi = 3.14
print('Pi:', pi)

# Boolean
is_active = True
is_valid = False
print('is_active:', is_active)
print('is_valid:', is_valid)

# None (like null in other languages)
result = None
print('result:', result)

# String
name = "John Doe"
print('Name:', name)

# List
numbers = [1, 2, 3, 4, 5]
print('Fruits:', numbers)

# Tuple
coordinates = (10, 20)
print('Coordinates:', coordinates)

# Set
fruits = {'apple', 'banana', 'orange'}
print('Fruits:', fruits)

# Dictionary
person = {'name': 'John', 'age': 25, 'city': 'New York'}
print(person)
```

And here comes some basic arithmetic operations (see interesting exponentiation operator):

```python
# Addition
sum = 4 + 6    		            #Result: 10

# Subtraction
difference = 10 - 3   	      #Result: 7

# Multiplication
product = 5 * 7		            #Result: 35

# Division
quotient = 17 / 4.0		        #Result: 4.25

# Modulo (remainder)
remainder = 17 % 4  	        #Result: 1

# Exponentiation
exponentiation = 2 ** 3    		#Result: 8

print("Sum: ", sum)
print("Subtraction: ", difference)
print("Multiplication: ", product)
print("Division: ", quotient)
print("Modulo: ", remainder)
print("Exponentiation: ", exponentiation)
```
